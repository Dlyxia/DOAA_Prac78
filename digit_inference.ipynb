```python
from PIL import Image
import requests
import numpy as np
# Load the image into arrays.
img = Image.open("digit_samples/img_6.jpg").convert("L")
img = img.resize((28, 28))
# Put them to correct shape so model can understand them.
img_array = np.array(img).reshape((1, 28, 28, 1)).astype("float32") / 255.0
img_array = img_array.tolist()
# Send the image to model service
response = requests.post(
"http://localhost:8501/v1/models/digit_classifier:predict",
json={
"instances": img_array
}
)
# Display the serverâ€™s response
predictions = response.json()
print("Predicted class:", np.argmax(predictions['predictions'][0]))
img #display the image
```
